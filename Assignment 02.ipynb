```python
#Tanmay Bhowmik; 19101465; CSE111 Section: 05
#CSE111 Assignment 2
#Easy 2

weight = int(input("Please enter your weight: "))
height = int(input("Please enter your height: "))

def bmi(weight , height):
    
    height /= 100
    
    bmi = weight / (height ** 2)
    bmi = round(bmi,1)
    
    if bmi<18.5:
        print("Score is " + bmi + ". " + "You are Underweight.")
    elif bmi>=18.5 and bmi<=24.9:
        print("Score is " + bmi + ". " + "You are Normal.")
    elif bmi>=25 and bmi<=30:
        print("Score is " + bmi + ". " + "You are Overweight.")
    else:
        print("Score is " + bmi + ". " + "You are Obese.")
    
bmi(weight ,height)        
```

```python
#Tanmay Bhowmik; 19101465; CSE111 Section: 05
#CSE111 Assignment 2
#Easy 3

minimum = int(input("Enter minimum value: "))
maximum = int(input("Enter maximum value: "))
divisor = int(input("Enter divisor: "))

def calc(minimum, maximum, divisor):
    
    sum = 0
    
    for i in range(minimum,maximum):
        if i%divisor == 0:
            sum += i
    print(sum)        
    return sum        

calc(minimum,maximum,divisor)
```

```python
#Tanmay Bhowmik; 19101465; CSE111 Section: 05
#CSE111 Assignment 2
#Medium 3

def vowel_count(full_name):
    
    vowels = ["a","e","i","o","u","A","E","I","O","U"]
    vowel = []
    count = 0
    
    for element in full_name:
        if element in vowels:
            vowel.append(element)
            count += 1
    print("Vowels: " , end = "")        
    print(*vowel , sep = ", " , end =".")
    print(" Total number of vowels: " + str(count) + ".")
        
vowel_count("My name is Tanmay.")  
```

```python
#Tanmay Bhowmik; 19101465; CSE111 Section: 05
#CSE111 Assignment 2
#Hard 1

from datetime import date
def employee_bonus_calculator(*a):
    argument_count = 0
    for i in a:
        argument_count+=1
    l = []
    for i in range(0,argument_count,3):
        name = a[i]
        bonus = a[i+1]
        joining_date = a[i+2]
        current_date = date.today()
        current_date = str(current_date)
        current_date = current_date.split("-")
        current_date = list(map(int, current_date))
        joining_date = joining_date.split("-")
        joining_date = list(map(int, joining_date))
        current_date = date(current_date[0],current_date[1],current_date[2])
        joining_date = date(joining_date[0],joining_date[1],joining_date[2])
        year = ((current_date-joining_date).days)/365
        if year<5:
            bonus = int(bonus*0.1)
        elif year>=5 and year<=10:
            bonus = int(bonus * 0.1 + 5000)
        else:
            bonus = int(bonus * 0.15 + 15000)
        l.append(name+": " + str(bonus))
    print(*l,sep = ", ")
        
employee_bonus_calculator('Alice', 20000, '2017-03-21', 'Bob', 50000, '2007-05-10') 
```

```python
#Tanmay Bhowmik; 19101465; CSE111 Section: 05
#CSE111 Assignment 2
#Hard 2

days = int(input("Please enter the number of days: "))

def remaining_years_months_days(days):
    
    years = days//365
    remainder = days - (years*365)
    months = remainder // 30
    days = days - (years * 365) - (months * 30)
    print(str(years) + ' years, ' + str(months) + " months and " + str(days) + " days")

remaining_years_months_days(days)
```

```python
#Tanmay Bhowmik; 19101465; CSE111 Section: 05
#CSE111 Assignment 2
#Hard 3

x = 'my favourite animal is a dog. a dog has sharp teeth so that it can eat flesh very easily. do you know my pet dog’s name? i love my pet very much.'

def capitalize(x):
    x = x.capitalize()
    x = x.split(" ")
    for i in range(len(x)):
       if i == len(x) - 1:
          break
       else:
         if x[i] == 'i':
            x[i] = x[i].capitalize()
         if '.'  in x[i]:
            x[i+1] = x[i+1].capitalize()
         if '?' in x[i]:
            x[i+1] = x[i+1].capitalize()
         if '!' in x[i]:
            x[i+1] = x[i+1].capitalize()
    print(*x , sep = " ") 
     
capitalize(x)
```
