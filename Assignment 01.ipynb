```python
#Assignment1 CSE111
#Tanmay Bhowmik; 19101465; CSE111 Section: 05
#medium problem3

number = int(input("Please enter a number: "))

def perfect_number_checker(number):
    sum_of_divisors = 0
    for i in range(1,number):
        if number%i == 0:
            sum_of_divisors += i
    if sum_of_divisors == number:
        print("Perfect Number")
    else:
        print("Not Perfect")
perfect_number_checker(number)
```

```python
#Assignment1 CSE111
#Tanmay Bhowmik; 19101465; CSE111 Section: 05
#hard problem2

num = int(input("Please enter a number: "))

for line_count in range(1,num+1):
    for space in range(line_count,num):
        print(" ", end = "")
    if line_count == num:
       for star in range(1,2*num):
           print("*",end="") 
    else:      
         for star in range (1,2*line_count):  
            if star == 1 or star == 2*(line_count)-1:
               print("*", end= "")
            else:
                 print(" ", end = "")
    print("")           
           
for line_count in range(1,num):
    for space in range(1,line_count+1):
        print(" ", end = "")
    for star in range(1,2*(num-line_count)):
        if star == 1 or star == 2*(num-line_count)-1:
            print("*",end="")
        else:
            print(" ", end="")
    print("")   
```
