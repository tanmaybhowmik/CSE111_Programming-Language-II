```python
#CSE111 Assignment 3
#Tanmay Bhowmik; 19101465; CSE111 Section: 05
#medium problem 1

x = input("Enter a string: ")
upper_case_count = 0 
lower_case_count = 0 

for i in range(len(x)):
    if x[i].isupper():
        upper_case_count += 1 
    elif x[i].islower():
        lower_case_count += 1 

if (upper_case_count > lower_case_count):
    print(x.upper())
else:
    print(x.lower())
```

```python
#CSE111 Assignment 3
#Tanmay Bhowmik; 19101465; CSE111 Section: 05
#medium problem 2 

x = input("Please enter a string: ")

digit_flag = False
string_flag = False

for i in range(len(x)):
    if x[i].isdigit():
        digit_flag = True
    else:
        string_flag = True
    if digit_flag == True and string_flag == True:
        break

if digit_flag == True and string_flag == False:
    print('NUMBER')
elif digit_flag == False and string_flag == True:
    print('WORD')
else:
    print('MIXED')
```

```python
#CSE111 Assignment 3
#Tanmay Bhowmik; 19101465; CSE111 Section: 05
#medium problem 3

x = input("Enter a string: ")
y = ''

for i in range(len(x)):
    if x[i].isupper():
        break
for j in range(i+1,len(x)):
    if x[j].isupper():
        break
    else:
        y += x[j]
        
if y == '':
    print('BLANK')
else:
    print(y)
```

```python
#CSE111 Assignment 3
#Tanmay Bhowmik; 19101465; CSE111 Section: 05
#hard problem 1

string1 = input("Please enter a string: ")
string2 = input("Please enter another string: ")
new_string = ''

for i in range(len(string1)):
    if string1[i] in string2:
        new_string += string1[i]
        
for i in range(len(string2)):
    if string2[i] in string1:
        new_string += string2[i]
        
if new_string == '':
    print('Nothing in common.')
else:
    print(new_string)
```

```python
#CSE111 Assignment 3
#Tanmay Bhowmik; 19101465; CSE111 Section: 05
# hard problem 2

s = input("Please enter a string: ")

for i in range(len(s)):
    if s.endswith(s[:i]):
        substring = s[:i]

if s.count(substring)>=3:
    print(substring)
else:
    print("Not Palindrome Substring")
```

```python
#CSE111 Assignment 3
#Tanmay Bhowmik; 19101465; CSE111 Section: 05
#hard problem 3

x = input("Please enter a string: ")
specials = ['_','$','#','@']

lower_case_missing = True
upper_case_missing = True
digit_missing = True
special_missing = True

for i in range(len(x)):
    if x[i].islower() == True:
        lower_case_missing = False
    if x[i].isupper() == True:
        upper_case_missing = False
    if x[i].isdigit() == True:
        digit_missing = False
    if x[i] in specials:
        special_missing = False

y = ['Lowercase character missing','Uppercase character missing','Digit missing','Special character missing','OK']
x = []

if lower_case_missing == True:
    x.append(y[0])
if upper_case_missing == True:
    x.append(y[1])
if digit_missing == True:
    x.append(y[2])
if special_missing == True:
    x.append(y[3])
if lower_case_missing == False and upper_case_missing == False and digit_missing == False and special_missing == False:
    x.append(y[4])

print(*x , sep = ", ") 
```
