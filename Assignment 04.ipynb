```python
#CSE111 Assignment 4
#Tanmay Bhowmik; 19101465; CSE111 Section: 05
# medium problem 1

input_numbers = []
num = input("Please enter your number or STOP: ")
while num != "STOP":
    input_numbers.append(num)
    num = input("Please enter your number or STOP: ")
freq_list = []
for i in range(len(input_numbers)):
    freq_list.append(str(input_numbers[i]) + ' - ' + str(input_numbers.count(input_numbers[i])) + ' times')
print(*sorted(set(freq_list)) , sep = "\n")
```

```python
#CSE111 Assignment 4
#Tanmay Bhowmik; 19101465; CSE111 Section: 05
# medium problem 2

num = int(input("Please enter the number of lists to be entered: "))

lst = []
while True:
    x = input("Please enter a number: ")
    if x == '':
        break
    else:
        lst.append(int(x))
s = sum(lst) 

fin_lst = []
for i in range(1,num):
    new_lists = []
    while True:
       x = input("Please enter a number: ")
       if x == '':
          break
       else:
          new_lists.append(int(x))
          if sum(new_lists)>s:
             s = sum(new_lists)
             fin_lst = new_lists

print(s)
print(fin_lst)
```

```python
#CSE111 Assignment 4
#Tanmay Bhowmik; 19101465; CSE111 Section: 05
# hard problem 1

lst = []
while True:
    x = input("Please enter a number: ")
    if x == 'STOP':
        break
    else:
        lst.append(int(x))

abs_diff_lst = []        
for i in range(len(lst)):
    if i == len(lst) - 1:
        break
    else:
        abs_diff_lst.append(abs(lst[i+1]-lst[i]))
        
abs_diff_lst.sort()

flag = False
for i in range(len(abs_diff_lst)):
    if abs_diff_lst[i] == i+1:
        flag = True
    else:
        flag = False
        
if flag == True:
    print('UB Jumper')
else:
    print('Not UB Jumper')
```

```python
#CSE111 Assignment 4
#Tanmay Bhowmik; 19101465; CSE111 Section: 05
# hard problem 2

string_input = input("Please enter your string: ")
def sort_numbers_in_string(string_input):
    string_input = "".join(sorted(string_input))
    y = ''
    string_num = ''
    for i in range(len(string_input)):
       if string_input[i].islower():
           y += string_input[i]
    for i in range(len(string_input)):
         if string_input[i].isupper():
           y += string_input[i]
    for i in range(len(string_input)):
        if string_input[i].isdigit():
           string_num += string_input[i]
    z = ''
    x = [int(x) for x in str(string_num)]
    for i in range(len(x)):
        if x[i]%2 == 1:
            y += str(x[i])
    for i in range(len(x)):
        if x[i]%2 == 0:
            z += str(x[i])
    y += z 
    return y  

print(sort_numbers_in_string(string_input))   
```

```python
#CSE111 Assignment 4
#Tanmay Bhowmik; 19101465; CSE111 Section: 05
# hard problem 3

num = int(input("Please enter number of potential participants: "))
k = int(input("Enter number of times you want each team to participate: "))

potential_participants = []
for i in range(num):
    potential_participants.append(int(input("Please enter the number of times he/she participated already: ")))

participants = []
for i in range(len(potential_participants)):
    if potential_participants[i]<=(5-k):
        participants.append(potential_participants[i])

if len(participants) < 3:
    print(0)
else:
    print(len(participants)//3)
```
